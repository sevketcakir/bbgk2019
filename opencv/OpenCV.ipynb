```python
import cv2
from matplotlib import pyplot as plt
```

```python
#Resimleri Jupyter notebook içinde göstermek için
def goster(img, baslik="Başlık yok"):
    plt.figure()
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(baslik)
    plt.show()
```

```python
img = cv2.imread('resimler/lenna.png')
```

```python
img.shape
```

```python
goster(img)
```

```python
messi = cv2.imread('resimler/messi.jpg')
goster(messi)
```

```python
messi=cv2.imread('resimler/messi.jpg')
ronaldo=cv2.imread('resimler/ronaldo.jpg')
ronaldo=ronaldo[0:360,0:620]
ronaldo=cv2.flip(ronaldo,1)
print(messi.shape, ronaldo.shape)
blend=cv2.addWeighted(messi,0.5,ronaldo,0.5,0)
goster(blend)
```

```python
import numpy as np
lenna=cv2.imread('resimler/messi.jpg')
#ma = np.array([110,50,50])
#mu = np.array([130,255,255])
ma = np.array([0,59,50])
mu = np.array([50,173,255])
lenna=cv2.cvtColor(lenna,cv2.COLOR_BGR2HSV)
maske = cv2.inRange(lenna, ma, mu)
goster(maske)
messi2=cv2.bitwise_and(messi,messi,mask=maske)
goster(messi2)
```

```python
buyuk=cv2.resize(messi, None, fx=4, fy=4, interpolation=cv2.INTER_CUBIC)
goster(buyuk)
cv2.imwrite("resimler/buyukmessi.jpg",buyuk)
```

```python
poker=cv2.imread('resimler/poker.jpg')
poker=cv2.cvtColor(poker, cv2.COLOR_BGR2GRAY)
goster(poker)
_,th=cv2.threshold(poker,115,255,cv2.THRESH_BINARY)
goster(th)
thadaptivec = cv2.adaptiveThreshold(poker,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,45,2)
goster(thadaptivec)
thadaptivegauss = cv2.adaptiveThreshold(poker,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,51,2)
goster(thadaptivegauss)
```
